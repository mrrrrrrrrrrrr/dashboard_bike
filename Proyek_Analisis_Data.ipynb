## Exploratory Data Analysis (EDA)

### Explore ...

```python
warnings.filterwarnings("ignore")

# Korelasi antara variabel numerik pada day
correlation_matrix = day.corr()
fig = px.imshow(correlation_matrix)
fig.update_layout(title="Korelasi antara Variabel Numerik")
fig.show()
```

```python
# Distribusi variabel kategorikal

categorical_cols = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
for col in categorical_cols:
    fig = px.bar(day[col].value_counts().reset_index(), x='index', y=col)
    fig.update_layout(title=f'Distribusi {col}')
    fig.show()
```

```python
# Hubungan antara musim (season) dan jumlah sewa (cnt)

fig = px.box(day, x='season', y='cnt')
fig.update_layout(title='Hubungan antara Musim dan Jumlah Sewa')
fig.show()
```

```python
# Hubungan antara hari libur (holiday) dan jumlah sewa (cnt)

fig = px.box(day, x='holiday', y='cnt')
fig.update_layout(title='Hubungan antara Hari Libur dan Jumlah Sewa')
fig.show()
```

```python
# Hubungan antara cuaca (weathersit) dan jumlah sewa (cnt)

fig = px.box(day, x='weathersit', y='cnt')
fig.update_layout(title='Hubungan antara Cuaca dan Jumlah Sewa')
fig.show()
```

```python
# Scatter plot antara suhu (temp) dan jumlah sewa (cnt)

fig = px.scatter(day, x='temp', y='cnt', title='Scatter Plot antara Suhu dan Jumlah Sewa')
fig.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapa banyak sepeda sewaan yang digunakan pada hari libur selama musim panas pada tahun 2012?

```python
filtered_data = day[(day["yr"] == 1) & (day["season"] == 2)]

# Hitung jumlah total sewa sepeda (cnt)
total_sewa_sepeda = filtered_data["cnt"].sum()

print("Jumlah total sewa sepeda untuk tahun 2012 selama musim panas:", total_sewa_sepeda)
```

### Pertanyaan 2: Apa korelasi antara suhu dan jumlah peminjaman sepeda?

```python
fig = px.scatter(day, x="temp", y="registered", title="Korelasi antara suhu dan jumlah peminjaman sepeda")
fig.update_xaxes(title="Suhu (temp)")
fig.update_yaxes(title="Jumlah Peminjam Sepeda")
fig.show()
```

Pada grafik terlihat korelasi positif yang menunjukkan bahwa meningkatnya suhu (temp) berpengaruh juga terhadap meningkatnya jumlah peminjam sepeda
